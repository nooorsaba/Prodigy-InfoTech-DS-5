```python
columns_to_drop = ['accident_index', 'accident_reference', 'vehicle_reference', 'casualty_reference']
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
import seaborn as sns

sns.countplot(x='accident_year', hue='sex_of_casualty', data=data)
plt.xlabel('Accident Year')
plt.ylabel('Count of Casualties')
plt.title('Casualties per Year by Gender')
plt.legend(title='Sex of Casualty')
plt.show()
```

```python
# Plotting road conditions using a bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='pedestrian_road_maintenance_worker', data=data)
plt.title('Road Conditions Distribution')
plt.xlabel('Road Conditions')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting accident hotspots on a map
plt.figure(figsize=(10, 8))
plt.scatter(data['car_passenger'], data['pedestrian_location'], alpha=0.1, s=5)
plt.xlabel('car_passenger')
plt.ylabel('pedestrian_location')
plt.title('Accident Hotspots')
plt.show()
```

```python
# Plotting contributing factors using a bar plot
plt.figure(figsize=(10, 6))
sns.countplot(y='accident_year', data=data, order = data['sex_of_casualty'].value_counts().index)
plt.title('Contributing Factors to Accidents')
plt.xlabel('Count')
plt.ylabel('Contributing Factors')
plt.show()
```
